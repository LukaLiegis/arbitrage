```python
import json
import websocket
import pandas as pd
from crypto_arbitrage import config
from crypto_arbitrage.data.binance_endpoints import BinanceEndpoints
```

```python
asset = config.ticker.lower() + "/ws/@bookTicker"
print(asset)
```

```python
socket = "wss://fstream.binance.com" + asset
print(socket)
```

```python
def manipulation(source):
    binance_bid = source["b"]
    #binance_ask = source["a"]
    #spread_binance = binance_ask - binance_bid
    binance_evt_time = source["E"]
    df = pd.DataFrame(binance_bid, columns=["binance_bid"], index=[binance_evt_time])
    df.index.name = "timestamp"
    df = df.astype(float)
    df = df.reset_index()
    print(df)
    return df
```

```python
def on_message(ws, message):
    message = json.loads(message)
    manipulation(message)
```

```python
ws = websocket.WebSocketApp(socket, on_message=on_message)
ws.run_forever()
```
